```python
#chatgpt
import logging
from abc import ABC, abstractmethod
from datetime import datetime, timedelta
import re


class BaseModule(ABC):
    def __init__(self, user_name):
        """Initialize the logger with a unique file for each session."""
        self.user_name = user_name
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        log_file = f"{self.user_name}_{timestamp}.log"

        # Set up a unique logger for this user
        self.logger = logging.getLogger(self.user_name)
        self.logger.setLevel(logging.INFO)

        # Create a file handler with the unique log file
        file_handler = logging.FileHandler(log_file)
        formatter = logging.Formatter("%(asctime)s - %(levelname)s - %(message)s")
        file_handler.setFormatter(formatter)

        # Add the handler to the logger  
        self.logger.addHandler(file_handler)

    @abstractmethod
    def input_data(self):  
        pass

#cgpt
# Specialization Module
class Specialization(BaseModule):
    def __init__(self, user_name):
        super().__init__(user_name)
        self.specializations = []

    def input_data(self):
        specialization_name = input("Enter Your Specialization: ").strip()
        try:
            if all(char.isalpha() or char.isspace() for char in specialization_name):
                self.specializations.append(specialization_name)
                self.logger.info("Specialization added: %s", specialization_name)
            else:
                raise ValueError("Only alphabetical characters allowed.")
        except ValueError as e:
            self.logger.error(e)

#cgpt
# Doctor Module
class Doctor(BaseModule):
    def __init__(self, user_name):
        super().__init__(user_name)
        self.license_no = None
        self.name = None
        self.phone = None
        self.email = None

    def input_data(self):
        self.license_no = self.get_license_no()
        self.name = self.get_name()
        self.phone = self.get_phone()
        self.email = self.get_email()

    def get_license_no(self):
        ls_no = input("Enter License No (Format: NNXXXXNN): ").strip()
        pattern = "^[0-9]{2}[A-Z]{4}[0-9]{2}$"
        try:
            if re.match(pattern, ls_no):
                self.logger.info("License No saved: %s", ls_no)
                return ls_no
            else:
                raise ValueError("Invalid License No format.")
        except ValueError as e:
            self.logger.error(e)
            return None

    def get_name(self):
        doc_name = input("Enter Your Name: ").strip()
        try:
            if all(char.isalpha() or char.isspace() for char in doc_name):
                self.logger.info("Doctor name saved: %s", doc_name)
                return doc_name
            else:
                raise ValueError("Invalid name format.")
        except ValueError as e:
            self.logger.error(e)
            return None

    def get_phone(self):
        doc_no = input("Enter Your Contact Number: ").replace(" ", "")
        pattern = "^[0-9]{10}$"
        try:
            if re.match(pattern, doc_no):
                self.logger.info("Phone number saved: %s", doc_no)
                return doc_no
            else:
                raise ValueError("Invalid phone number format.")
        except ValueError as e:
            self.logger.error(e)
            return None

    def get_email(self):
        mail = input("Enter Your Email ID: ").strip()
        pattern = r"^[^@]+@[a-z]+\.[a-z]+$"
        try:
            if re.match(pattern, mail):
                self.logger.info("Email saved: %s", mail)
                return mail
            else:
                raise ValueError("Invalid email format.")
        except ValueError as e:
            self.logger.error(e)
            return None

#cgpt
# Example Usage
if __name__ == "__main__":
    user_name = input("Enter your username: ").strip()
    doctor = Doctor(user_name)
    doctor.input_data()
```

```python
class IPD(BaseModule):
    """Handles admission and discharge details along with foreign keys."""
    def _init_(self):
        super()._init_()
        self.admission_id = None  
        self.admission_date = None
        self.discharge_date = None
        self.medicine_id = None 
        self.patient_id = None  
        self.doc_ls_no = None  

    def input_admission_data(self):
        self.admission_date = self.get_admission_date()
        self.discharge_date = self.get_discharge_date()

    def get_admission_date(self):
        """Gets and validates the admission date."""
        admn_date = input("Enter Admission Date (YYYY-MM-DD): ").strip()
        try:
            admission_date = datetime.strptime(admn_date, "%Y-%m-%d")
            if admission_date <= datetime.today():
                self.logger.info("Admission date saved: %s", admission_date)
                return admission_date
            else:
                raise ValueError("Admission date cannot be in the future.")
        except ValueError as e:
            self.logger.error(e)
            print(e)
            return None

    def get_discharge_date(self):
        """Gets and validates the discharge date."""
        disch_date = input("Enter Discharge Date (YYYY-MM-DD): ").strip()
        try:
            discharge_date = datetime.strptime(disch_date, "%Y-%m-%d")
            if (
                self.admission_date
                and discharge_date <= datetime.today()
                and discharge_date >= self.admission_date
            ):
                self.logger.info("Discharge date saved: %s", discharge_date)
                return discharge_date
            else:
                raise ValueError(
                    "Discharge date cannot be in the future and must be after the admission date."
                )
        except ValueError as e:
            self.logger.error(e)
            print(e)
            return None
```
