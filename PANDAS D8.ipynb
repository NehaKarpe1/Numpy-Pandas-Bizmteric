```python
import pandas as pd
l1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
r1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K30'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})  
```

```python
 
r2 = pd.DataFrame({'key': ['K3', 'K5', 'K6', 'K30'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})  
```

```python
pd.merge(l1,r1,how='inner',on='key')
```

```python
pd.merge(l1,r1,how='left',on='key')
```

```python
pd.merge(l1,r1,how='right',on='key')
```

```python
d1=pd.merge(l1,r2,how='outer',on='key')
d1
```

```python
pd.merge(d1,r1,how='inner',on='key')  
```
